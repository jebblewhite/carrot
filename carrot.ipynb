```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("carrot_data.csv")

df
```

```python
df.info()
```

```python
df = df.loc[df['Value'] > 0]
```

```python
df_prod = df.loc[df['Element'] == 'Production']
```

```python
productionval = df_prod["Value"]
sns.displot(productionval)
plt.show()
```

```python
productionval.describe()
```

```python
china_prod_yrs = df_prod.loc[df['Area' ] == 'China']
```

```python
china_prod_yrs
```

```python
plt.plot(china_prod_yrs['Year'],china_prod_yrs['Value'])
plt.show()
```

```python
for country in df_prod['Area']:
    values = df_prod.loc[df_prod['Area'  == country]]
    plt.plot(df_prod['Year'],values)
    plt.show()
```
